```python
#Models Import
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#Q1. Downloading the data
#JAN - https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet
#FEB - https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet')
```

```python
#to see the data headers
df.head()
```

```python
#read the columns
len(df.columns)
```

```python
#Q2. Computing duration and adding duration column
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
df['duration'] = df.duration.dt.total_seconds() / 60
df.head()
```

```python
#find the avergae of all the rides
df.duration.mean()
```

```python
#find the standard deviation of all the rides duration
df.duration.std()
```

```python
len(df.duration)
```

```python
#Q3. Dropping outliers
#count of the rides between 1 and 60 mins
len(df[(df.duration >= 1) & (df.duration <= 60)])
```

```python
len(df[(df.duration >= 1) & (df.duration <= 60)]) / len(df) * 100
```

```python
df = df[(df.duration >= 1) & (df.duration <= 60)].copy()
```

```python
len(df.duration)
```

```python
#Q4. One-hot encoding
categorical = ['PULocationID', 'DOLocationID']
```

```python
df[categorical] = df[categorical].astype(str)
```

```python
train_dicts = df[categorical].to_dict(orient='records')
```

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
```

```python
print(f'Feature matrix size: {X_train.shape}')
```

```python
#Q5. Training a model
target = 'duration'
y_train = df[target].values
```

```python
print (y_train)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_train)

import numpy as np
print(f'Train RMSE: {np.sqrt(mean_squared_error(y_train, y_pred))}')
```

```python
sns.histplot(y_pred, label='prediction')
sns.histplot(y_train, label='actual')

plt.legend();
```
